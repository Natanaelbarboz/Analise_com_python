```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("C:/Users/Natanael/Desktop/BOOTCAMP_UNIMED-BH/PYTHON/datasets/Aracaju.xlsx")
df2 = pd.read_excel("C:/Users/Natanael/Desktop/BOOTCAMP_UNIMED-BH/PYTHON/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("C:/Users/Natanael/Desktop/BOOTCAMP_UNIMED-BH/PYTHON/datasets/Natal.xlsx")
df4 = pd.read_excel("C:/Users/Natanael/Desktop/BOOTCAMP_UNIMED-BH/PYTHON/datasets/Recife.xlsx")
df5 = pd.read_excel("C:/Users/Natanael/Desktop/BOOTCAMP_UNIMED-BH/PYTHON/datasets/Salvador.xlsx")
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df["Dia_Venda"], df["MÃªs_Venda"] = (df["Data"].dt.day, df["Data"].dt.month)
```

```python
df["Trimestre_Venda"] = df["Data"].dt.quarter
```

```python
df.groupby("Cidade") ["Vendas"].sum()
```

```python
df.groupby("Cidade") ["Vendas"].sum().plot.bar(title = "Vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Valor Vendas")
plt.xticks(rotation = "horizontal");
```

```python
df.groupby("MÃªs_Venda") ["Vendas"].sum()
```

```python
df.groupby("MÃªs_Venda") ["Vendas"].sum().plot.bar(title = "Vendas por mÃªs")
plt.xlabel("Meses")
plt.ylabel("Valor Vendas")
plt.xticks(rotation = "horizontal");
```

```python
df.groupby("Trimestre_Venda") ["Vendas"].sum()
```

```python
df.groupby("Trimestre_Venda") ["Vendas"].sum().plot.bar(title = "Vendas por Semestre")
plt.xlabel("Semestre")
plt.ylabel("Valor Vendas")
plt.xticks(rotation = "horizontal");
```

```python
df.groupby("Ano_Venda") ["Vendas"].sum()
```

```python
df.groupby("Ano_Venda") ["Vendas"].sum().plot.bar(title = "Vendas por Ano")
plt.xlabel("Anos")
plt.ylabel("Valor Vendas")
plt.xticks(rotation = "horizontal");
```
